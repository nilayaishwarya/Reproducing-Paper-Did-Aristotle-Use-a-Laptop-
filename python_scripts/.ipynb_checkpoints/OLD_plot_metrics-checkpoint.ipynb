```python
import json
import os
import re
```

```python
paths = ['/home/jupyter/src/strategyqa/output_ir-ora-d/', '/home/jupyter/src/strategyqa/output_ir-ora-d_hyper0/', '/home/jupyter/src/strategyqa/output_ir-ora-d_hyper1/']#,'/home/jupyter/src/strategyqa/output_ir-ora-d_hyper2/','/home/jupyter/src/strategyqa/output_ir-ora-d_hyper3/']
```

```python
def load_metrics(p, id):
    file = open(p)
    data = json.load(file)
    return data
    
def print_metric(id, metric):
    print(id)
    print(metric['validation_accuracy'])
    print(metric['validation_loss'])
```

```python
def load_metrics_from_path(path):
    metrics = []
    paths = os.listdir(path)
    metrics_paths = {}
    for p in paths:
        match = re.search("metrics_epoch_(.*).json", p)
        if match:
            id = int(match.group(1))
            metrics_paths[id] = path+p

    for id in sorted(metrics_paths):
        metric = metrics_paths[id]
        metrics.append(load_metrics(metric, id))
        #print_metric(id, metrics[id])
    return metrics
```

```python
metrics = []
for path in paths:
    metrics.append({'label': os.path.basename(os.path.normpath(path)), 'data': load_metrics_from_path(path)})
```

```python
from pylab import *

def plot_metrics: 
    for metric in metrics:
        plot([m['validation_loss'] for m in metric['data']], label=metric['label'])

    legend()
    grid(True)
    xlabel('Epoch')
    ylabel('Value')
    title('Validation loss')
    show()

    for metric in metrics:
        plot([m['validation_accuracy'] for m in metric['data']], label=metric['label'])

    # plot(t, s2)

    xlabel('Epoch')
    ylabel('Value')
    ylim(ymin=0.55)
    title('Validation accuracy')
    legend()
    grid(True)
    show()

    for metric in metrics:
        plot([m['training_loss'] for m in metric['data']], label=metric['label'])

    xlabel('Epoch')
    ylabel('Value')
    title('Training loss')
    legend()
    grid(True)
    show()

    for metric in metrics:
        plot([m['training_accuracy'] for m in metric['data']], label=metric['label'])

    xlabel('Epoch')
    ylabel('Value')
    title('Training accuracy')
    legend()
    grid(True)
    show()
```
